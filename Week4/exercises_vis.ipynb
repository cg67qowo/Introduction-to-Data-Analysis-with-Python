```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/iris")

print("Path to dataset files:", path)
filepath=path+"/Iris.csv"
print(filepath)
```

```python
with open(filepath, 'r') as f:
    df = pd.read_csv(f)
df.head()
```

```python
# Build a scatterplot of petal length per species

colors = {'Iris-setosa':'red', 'Iris-versicolor':'green', 'Iris-virginica':'blue'}

plt.figure(figsize=(8, 6))

# Loop over each species to plot points
for species in df['Species'].unique():
    subset = df[df['Species'] == species]
    plt.scatter(
        [species]*len(subset),  # x-values: repeat species name
        subset['PetalLengthCm'], # y-values
        color=colors[species],
        alpha=0.7,
        s=60,
        label=species
    )

plt.title('Scatter Plot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()

```

```python
# Group unique species with the average of the petal length

species=df['Species'].unique()
print(species)

grouped = df.groupby('Species')['PetalLengthCm'].mean()
grouped = grouped.reset_index() # By default, groupby delivers a Series. This commands convert the Series to a DataFrame. 
# If you are not confident about what is being done here, you can revise the concepts of "Series" and "DataFrame" by looking at last week's notebooks.
grouped.head()
```
