#### 2. Data Cleaning
We saw that the dataset has some missing values. Let's handle them.
The method `isnull` prints `True` if the value is missing, `False` otherwise.

```python
df["Province/State"].isnull()
```

We see that the column `Province/State` contains several missing values. We can print out the amount of missing values with `value_counts()`:

```python
df["Province/State"].isnull().value_counts()
```

`dropna` removes the rows with missing values. Let's do it for the `Province/State` column.

```python
df_clean_province_values = df["Province/State"].dropna()
df_clean_province_values.head()
```

Good, but we need to see the whole df. Let's apply the same method to the `Province/State` column, but for the whole dataset.

```python
df_clean_province_values = df.dropna(subset=['Province/State'])
df_clean_province_values.head()
```

An alternative could be filling the missing values with 0.

```python
df_zeros=df.fillna(0)  # Replace missing values with 0
df_zeros.head()
```

```python
# If you want to modify the dataframe directly, run:
#df.fillna(0, inplace=True)
```

In case of **numerical values**, replacing the item with **the mean** is also very useful: `df.fillna(df.mean())`

Part of data cleaning is also **choosing the right strategy**. For example, another method is **filling with forward propagation**: the value previous to the null value in the column will be used to fill the null value. Similarly, **backward propagation** will mean that the value after the null value is used to fill the null value. In our case, `ffill` cannot be applied, as the first row already contains a NaN value.

```python
df_backward=df.bfill()
df_backward.head()
```

Clearly, this method is **not useful in this context**, as it would return values that are simply wrong. You can find more information about techniques to handle missing values in the dedicated dataset. However, choosing the technique will mostly depend on your own data and research question, as more elaborate statistical methods are not into the scope of this class. You will be asked to choose based on the properties of your data: For example, replacing NaNs with 0 in continuous variables (like height, temperature, income) can artificially lower the mean and inflate variance, which would be bad.

##### Grouping and aggregating
`Pandas` offers useful ways of grouping data, e.g. by using the `groupby` operator. In the following example, you will see how to group the dataset by WHO region:

```python
df_grouped = df.groupby('WHO Region')
df_grouped.get_group('Europe')
```

The previous example **groups specific rows**, but does not perform any operation on them. When operations are performed in addition to the grouping, we talk of **aggregation**. Below is an example of aggregation: we sum the amount of confirmed COVID cases per each WHO region.

```python
confirmed_cases_df = df.groupby('WHO Region')['Confirmed'].sum()
confirmed_cases_df
```

We will now compute the average of confirmed cases and death cases per WHO region:

```python
df.groupby('WHO Region')[['Deaths', 'Confirmed']].mean()
```

The two operations (`sum` and `mean`) can be performed together, with the help of `agg()`. Try doing it yourself.
> Hint: agg() takes as parameter a dictionary where the keys are the column names, and the values are the operations you want to perform, such as 'sum' or 'mean'.

```python
# Exercise: Group the dataset by WHO Region and display the sum of confirmed cases, and the mean of deaths.
```

```python
# Solution: Aggregation

df.groupby('WHO Region').agg({
    'Confirmed': 'sum',    # sum Quantity per category
    'Deaths': 'mean'       # average Price per category
})
```

##### Merging dataframes

The most widely used technique is the **inner merge**, that merges dataframes by keeping only rows present in both dataframes, and including all columns. Information about other kinds of merging can be found [here](https://pandas.pydata.org/docs/reference/api/pandas.merge.html).

```python
second_df = pd.DataFrame({
        'Population (estimate)': [
        1585000000,   # Africa
        1062000000,   # Americas
        745000000,    # Eastern Mediterranean
        743000000,    # Europe
        695000000,    # South-East Asia
        2200000000    # Western Pacific
    ],
    'WHO Region': ['Africa','Americas','Eastern Mediterranean','Europe', 'South-East Asia', 'Western Pacific']
})
```

```python
merged_inner = pd.merge(confirmed_cases_df, second_df, on='WHO Region', how='inner')
merged_inner
merged_inner.shape
```
